```python
import pandas as pd
import plotly.express as px
```

```python
# carregue o dataset
data = pd.read_csv("precos_gasolina_etanol_03.csv",sep=';')
```

```python
# transformar dataset em dataframe
df = pd.DataFrame(data)
```

```python
# imprimir linhas do dataset
df.head()
```

```python
# imprima as informações da base de dados
print(df.info())
```

```python
# verifique se há dados faltantes. se houver, remova os valores vazios
df.isnull()
```

```python
#removendo os valores vazios
df.dropna(axis=1)
```

```python
# verifique se há colunas que não são relevantes para uma anaálise inicial dos preços e se houver, exclua
df.drop(['CNPJ da Revenda', 'Municipio', 'Nome da Rua', 'Cep', 'Unidade de Medida'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# filtre os registros apenas do estado de Pernambuco (PE).
df[df['Estado - Sigla'] == 'PE']
```

```python
# crie um gráfico de barras mostrando o preço médio de "Valor de Venda" por estado.
bar_plot = px.bar(
          df, # Dataframe
          x='Estado - Sigla',  # Dados no eixo X
          y="Valor de Venda",  # Dados no eixo Y
          title="Vendas p/ Estado",  # Título do gráfico
          labels={'Estado - Sigla', 'Valor de Venda'},  # Rótulos dos eixos (dicionário)
          color="Estado - Sigla",    # Colorindo as barras com base nos valores
          template="plotly_white"  # Estilo do gráfico
)

bar_plot.show() # Exibindo o gráfico
```

```python
dic_produtos = df[df['Produto'] == 'GASOLINA'][['Data da Coleta', 'Valor de Venda']]
```

```python
dic_produtos
```

```python
dic_produtos['Valor de Venda'] = dic_produtos['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
dic_produtos['Data da Coleta'] = dic_produtos['Data da Coleta'].str.replace('/','-')
```

```python
df_produtos = pd.DataFrame(dic_produtos)
df_produtos
```

```python
soma_e_contagem = []
```

```python
for index, row in df_produtos.iterrows(): 
    data = row['Data da Coleta'] 
    valor = row['Valor de Venda'] 
    if data not in soma_e_contagem: 
        soma_e_contagem[data] = {'soma': 0, 'contagem': 0} 
    soma_e_contagem[data]['soma'] += valor 
    soma_e_contagem[data]['contagem'] += 1
```

```python
# crie um gráfico de linhas do preço médio de "Valor de Venda" ao longo do tempo para gasolina.
time_series = px.scatter(
              dic_produtos, # Dataframe
              x= "Data da Coleta",  # Eixo X
              y= "Valor de Venda",  # Eixo Y
              title="",
              labels={'Produto':'GASOLINA'},
              template="plotly_white"  # Tema limpo
)

# Exibindo o gráfico
time_series.show()
```
